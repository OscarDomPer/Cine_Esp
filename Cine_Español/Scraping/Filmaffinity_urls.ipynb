```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

# ---------------------- INICIALIZACIÓN ----------------------

# Intenta leer el archivo CSV para obtener las URLs ya procesadas.
# Si no existe el archivo, crea uno nuevo.
try:
    df_peliculas = pd.read_csv("data/data_peliculas.csv")
    urls_procesadas = df_peliculas['URL'].tolist()
except FileNotFoundError:
    urls_procesadas = []
    # Crear el archivo CSV con encabezados si no existe
    with open("data/data_pelis.csv", "w") as file:
        file.write("Nombre,URL\n")

url = "https://www.filmaffinity.com/es/topgen.php?genres=&chv=0&orderby=avg&movietype=movie%7C&country=ES&fromyear=1900&toyear=2023&ratingcount=1&runtimemin=0&runtimemax=4"

peliculas_info = []

# ---------------------- EXTRACCIÓN DE DATOS ----------------------

driver = webdriver.Chrome()  # O el navegador que estés utilizando
driver.get(url)

while True:
    soup = BeautifulSoup(driver.page_source, "html.parser")
    movies = soup.find("main", id="mt-content-cell").find_all("div", class_="mc-title")

    for movie in movies:
        title = movie.find("a").get_text(strip=True)
        link = movie.find("a")["href"]

        # Si la URL ya ha sido procesada, la omitimos
        if link in urls_procesadas:
            continue

        print(f"Procesando {title}: {link}")

        peliculas_info.append({
            "Nombre": title,
            "URL": link
        })

        # Guardar en el CSV cada 10 URLs para no perder datos en caso de fallos
        if len(peliculas_info) % 10 == 0:
            df_temp = pd.DataFrame(peliculas_info)
            df_temp.to_csv("data/data_pelis.csv", mode='a', header=False, index=False)
            peliculas_info = []  # Limpiar la lista para evitar duplicados

        #time.sleep(random.uniform(1, 5))  # Esperar un tiempo aleatorio entre 1 y 5 segundos

    try:
        load_more_button = driver.find_element_by_class_name("fas fa-chevron-down")
        load_more_button.click()
        time.sleep(2)  # Esperar a que se cargue el contenido
    except:
        break  # Si no hay más botones para cargar más contenido, salir del bucle

#driver.quit()

# ---------------------- GUARDAR DATOS RESTANTES ----------------------

# Guardar las URLs restantes en el CSV
if peliculas_info:
    df_temp = pd.DataFrame(peliculas_info)
    df_temp.to_csv("data/data_pelis.csv", mode='a', header=False, index=False)

# Mostrar el DataFrame resultante
df_peliculas = pd.read_csv("data/data_pelis.csv")
df_peliculas
```
