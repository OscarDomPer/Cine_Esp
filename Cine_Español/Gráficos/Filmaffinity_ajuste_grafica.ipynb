```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargamos los datos desde el archivo CSV
ruta_datos_peliculas = 'data/data_peliculas_transformado_filtrado.csv'

# Leemos el archivo CSV en un DataFrame de pandas
df_peliculas = pd.read_csv(ruta_datos_peliculas)

# Convertimos la columna 'Año' a tipo datetime para poder extraer el año
df_peliculas['Año'] = pd.to_datetime(df_peliculas['Año'], errors='coerce').dt.year

# Agrupamos las películas por año y contamos la cantidad de películas por cada año
peliculas_por_año = df_peliculas.groupby(df_peliculas['Año'])['Título'].count()

# Definimos la función para añadir el sombreado de fondo y las líneas divisorias con anotaciones verticales centradas
def add_centered_background_shading_and_dividers(plot):
    # Definir los períodos y sus etiquetas
    periods = [
        (1900, 1936, 'Etapa inicial: 1900 a 1936'),
        (1936, 1960, 'Primera etapa del Franquismo: 1936 a 1960'),
        (1960, 1975, 'Segunda Etapa del Franquismo: 1960 a 1975'),
        (1975, 1983, 'Destape: 1975 a 1983'),
        (1983, 2007, 'Periodo: 1983 a 2007'),
        (2007, 2020, 'Crisis Financiera: 2007 a 2020'),
        (2020, 2023, 'Pandemia Covid: 2020 a 2023')
    ]
    
    # Colores para el sombreado de fondo
    colors = ['#DDEEFF', '#AACCFF', '#77AAFF', '#4488FF', '#1166FF', '#0044FF', '#0022BB']
    
    # Obtenemos los límites actuales del eje Y
    ylim = plot.get_ylim()
    vertical_center = (ylim[0] + ylim[1]) / 2
    
    # Aplicar el sombreado de fondo y las anotaciones centradas
    for i, (start, end, label) in enumerate(periods):
        color = colors[i % len(colors)]
        plot.axvspan(start, end, color=color, alpha=0.1)
        plot.text((start + end) / 2, vertical_center, label, ha='center', va='center', 
                  rotation='vertical', color='grey', alpha=0.7)

# Creamos una gráfica de serie temporal con la cantidad de películas por año
plt.figure(figsize=(15, 7))
plot = sns.lineplot(x=peliculas_por_año.index, y=peliculas_por_año.values, color='steelblue', marker='o')
plt.title('Número de Películas por Año en el Dataset', fontsize=15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Número de Películas', fontsize=12)
plt.grid(True)

# Aplicar sombreado de fondo y líneas divisorias centradas
add_centered_background_shading_and_dividers(plot)

# Añadimos la línea temporal original con puntos marcadores
plot.plot(peliculas_por_año.index, peliculas_por_año.values, marker='o', color='steelblue')

plt.show()
```
