```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame
ruta_archivo = 'data/data_peliculas_transformado.csv'
datos = pd.read_csv(ruta_archivo)

# Función para obtener el primer actor o actriz de la lista de reparto
def obtener_protagonista(reparto):
    if isinstance(reparto, str):
        reparto_lista = eval(reparto)
        if reparto_lista:
            return reparto_lista[0]
    return ''

# Crear la nueva columna 'Protagonistas'
datos['Protagonistas'] = datos['Reparto'].apply(obtener_protagonista)

# Mostrar las primeras filas del DataFrame
datos.head()
```

```python
ruta_archivo_salida = "data/data_peliculas_transformado.csv"
datos.to_csv(ruta_archivo_salida, index=False)
```
