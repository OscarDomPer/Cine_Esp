```python
import pandas as pd
import ast

# Cargar el archivo CSV en un DataFrame
ruta_archivo = '"data/data_peliculas_transformado_filtrado.csv"'
df = pd.read_csv(ruta_archivo)

# Convertir la representación de la lista en string a una lista real
df['Reparto'] = df['Reparto'].apply(ast.literal_eval)

# Extraer actores únicos de la columna 'Reparto'
actores_unicos = set()
for reparto in df['Reparto']:
    actores_unicos.update(reparto)

# Convertir set a lista para poder ordenarla
actores_unicos = list(actores_unicos)
actores_unicos.sort()

# Crear un DataFrame para cada actor
lista_peliculas_actor = []
for actor in actores_unicos:
    peliculas_actor = df[df['Reparto'].apply(lambda x: actor in x)]['Título'].tolist()
    lista_peliculas_actor.append([actor, peliculas_actor, None])  # None para la columna Sexo

# Crear el DataFrame final
df_actores = pd.DataFrame(lista_peliculas_actor, columns=['Actor', 'Peliculas', 'Sexo'])

# Guardar el DataFrame en un archivo CSV
df_actores.to_csv('data/data_actores_filmaffinity.csv', index=False)
df_actores
```

```python
import pandas as pd
import gender_guesser.detector as gender
from tqdm import tqdm

# Cargar el archivo CSV en un DataFrame
ruta_archivo = 'data/data_actores_filmaffinity.csv'
df = pd.read_csv(ruta_archivo)

# Crear un detector de género
detector = gender.Detector()

# Función para obtener el género del nombre
def obtener_genero(nombre):
    primer_nombre = nombre.split()[0] if nombre.split() else nombre
    genero = detector.get_gender(primer_nombre)
    if genero == 'male' or genero == 'mostly_male':
        return 'Hombre'
    elif genero == 'female' or genero == 'mostly_female':
        return 'Mujer'
    else:
        return ''

# Aplicar la función obtener_genero para actualizar la columna 'Sexo' en el DataFrame
tqdm.pandas()
df['Sexo'] = df['Actor'].progress_apply(obtener_genero)

# Guardar el DataFrame actualizado en un archivo CSV
nombre_archivo_salida = 'data/data_actores_filmaffinity.csv'
df.to_csv(nombre_archivo_salida, index=False)

# Mostrar mensaje de confirmación
print(f"El DataFrame ha sido actualizado y guardado en un archivo CSV llamado '{nombre_archivo_salida}'")

# Mostrar el DataFrame
df
```
